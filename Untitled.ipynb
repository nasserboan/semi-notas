Encontrar grupos de alunos que possam ser auxiliados de forma sistêmica e não individualizada, ou seja, identificar grupos de alunos que possuem uma perfomance abaixo da média em um grupo de matérias, entender sua distribuição nas escolas do DF e recomendar um plano de otimização de aulas de reforço.

```python
import pandas as pd
import tqdm

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## selecionando os tipos de algumas colunas
converters= {'cod_escola':str,'cod_serie':str,'cod_turno':str,'cod_aluno':str,'cod_disciplina':str}

## lendo o arquivo csv
df = pd.read_csv('data/filtered.csv',dtype=converters)
```

```python
df[df.nota_bimestral > 10].shape
```

```python
def correct_nota(value):
    if value > 10:
        return round(value/10,2)
    else:
        return value
```

```python
df['nota_corrigida'] = df['nota_bimestral'].apply(correct_nota)
```

```python
df[df.nota_corrigida > 10].shape
```
